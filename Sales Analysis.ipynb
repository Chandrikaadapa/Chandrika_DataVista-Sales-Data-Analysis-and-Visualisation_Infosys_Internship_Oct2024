#### DRIVE YEAR AND MONTH FROM 'order_date' FOR FUTURE ANALYSIS

```python
sales['month_year'] = sales['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
sales['month_year'].sample(10)
```

## EXPLORATORY ANALYSIS

```python
sales.head()
```

#### GROUP SALES BY MONTHS 

```python
# Group by 'month_year' and sum only the numeric columns
sales_by_month = sales.groupby('month_year').sum(numeric_only=True).reset_index()
```

### VISUALIZE SALES TRAND BY MONTHS

```python
plt.figure(figsize=(15,6))
plt.plot(sales_by_month['month_year'],sales_by_month['sales'])
plt.xticks(rotation='vertical',size=8)
plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.bar(sales_by_month['month_year'],sales_by_month['sales'])
plt.xticks(rotation='vertical',size=8)
plt.show()
```

### DISPLAY MOST SELLING PRODCUTS

```python
products_sales = pd.DataFrame(sales.groupby('product_name')['sales'].sum())
products_sales = products_sales.sort_values('sales', ascending=False)
```

### TOP 10 MOST SALES PRODUCTS 

```python
products_sales[:10]
```

```python
products_by_quantity = sales.groupby('product_name')['quantity'].sum()
products_by_quantity_sorted = products_by_quantity.sort_values(ascending=False)
```

#### TOP 10 MOST QUANTITY SELLING PRODUCTS ITEMS

```python
products_by_quantity_sorted[:10]
```

#### VISUALIZE MOST USED SHIP MODS

```python
plt.figure(figsize=(10,8.5))

sns.countplot(x='ship_mode', data=sales)
plt.show()
```

### Distribution of Shiping Modes

```python
ship_mode = pd.DataFrame(sales['ship_mode'].value_counts()).reset_index()

ship_mode = ship_mode.rename({'index': 'shiping_mode', 'ship_mode': 'count'}, axis=1)

ship_mode.head()
```

```python
fig = plt.figure(figsize =(10, 7))
plt.pie(ship_mode['count'], labels=ship_mode['shiping_mode'])
# show plot
plt.show()
```

### MOST SELLING SEGMENTS

```python
plt.figure(figsize=(10,8.5))
sns.countplot(data=sales, x='segment')
plt.title('Count of Sales by Segment')  # Optional: Add a title for clarity
plt.xlabel('Segment')  # Optional: Label for the x-axis
plt.ylabel('Count')
plt.show()
```

### SEGMENTS BY SELLING QUANTITY

```python
# Group by 'segment' and sum only the 'quantity' column
by_segment = sales.groupby('segment')['quantity'].sum().reset_index()

# Sort the resulting DataFrame by 'quantity' in descending order
by_segment_sort = by_segment.sort_values('quantity', ascending=False)

# Display the sorted DataFrame
print(by_segment_sort)
```

### SEGMENTS BY PROFIT

```python
# Group by 'segment' and sum only the 'profit' column
by_segment_p = sales.groupby('segment')['profit'].sum().reset_index()

# Sort the resulting DataFrame by 'profit' in descending order
by_segment_sort_p = by_segment_p.sort_values('profit', ascending=False)

# Display the sorted DataFrame
print(by_segment_sort_p)
```

### SUB - CATEGORIES BY SHIPING COST

```python
# Ensure that only numeric columns are summed, ignoring non-numeric data
by_shipping_cost = sales.groupby('sub_category')['shipping_cost'].sum().reset_index()

# Sort the DataFrame by shipping cost in descending order
by_shipping_cost_sort = by_shipping_cost.sort_values('shipping_cost', ascending=False)

# Display the sorted DataFrame
by_shipping_cost_sort
```

### DISCOUNTS BY SUB - CATEGORIES

```python
# Ensure that only numeric columns are summed
by_discount = sales.groupby('sub_category')['discount'].sum().reset_index()

# Sort the DataFrame by discount in descending order
by_discount_sort = by_discount.sort_values('discount', ascending=False)

# Display the sorted DataFrame
by_discount_sort
```
